```python
def done(mat):
    for i in range(9):
        for j in range(9):
            if mat[i][j] == 0:
                return False
    return True
```

```python
auxMat =  [[[],[],[],[],[],[],[],[],[]],[[],[],[],[],[],[],[],[],[]],[[],[],[],[],[],[],[],[],[]],[[],[],[],[],[],[],[],[],[]],[[],[],[],[],[],[],[],[],[]],[[],[],[],[],[],[],[],[],[]],[[],[],[],[],[],[],[],[],[]],[[],[],[],[],[],[],[],[],[]],[[],[],[],[],[],[],[],[],[]]]
```

```python
def startMat(mat):
    for i in range(9):
        for j in range(9):
            if mat[i][j] != 0:
                auxMat[i][j].append(-1)
```

```python
def notInColumn(mat, col, num):
    for i in range(9):
        if mat[i][col] == num:
            return False
    return True
```

```python
def notInRow(mat, row, num):
    for i in range(9):
        if mat[row][i] == num:
            return False
    return True
```

```python
def notInGrid(mat, row, col, num): 
    if row == 0 or row == 3 or row == 6:
        if col == 0 or col == 3 or col == 6: #top left
            for i in range(1,3):
                for j in range(1,3):
                    if mat[row+i][col+j] == num:
                        return False
            return True

        elif col == 1 or col == 4 or col == 7: #top middle
            for i in range(1,3):
                if mat[row+i][col+1] == num:
                    return False
                if mat[row+i][col-1] == num:
                    return False
            return True

        elif col == 2 or col == 5 or col == 8: #top right
            for i in range(1,3):
                for j in range(1,3):
                    if mat[row+i][col-j] == num:
                        return False
            return True

    elif row == 1 or row == 4 or row == 7:
        if col == 0 or col == 3 or col == 6: #middle left
            for j in range(1,3):
                if mat[row+1][col+j] == num:
                    return False
                if mat[row-1][col+j] == num:
                    return False
            return True
          
            
        elif col == 1 or col == 4 or col == 7: #middle middle
            if mat[row+1][col+1] == num:
                return False
            if mat[row-1][col+1] == num:
                return False
            if mat[row+1][col-1] == num:
                return False
            if mat[row-1][col-1] == num:
                return False
            return True

        elif col == 2 or col == 5 or col == 8: #middle right
            for j in range(1,3):
                if mat[row+1][col-j] == num:
                    return False
                if mat[row-1][col-j] == num:
                    return False
            return True

    elif row == 2 or row == 5 or row == 8:
        if col == 0 or col == 3 or col == 6: #bottom left
            for i in range(1,3):
                for j in range(1,3):
                    if mat[row-i][col+j] == num:
                        return False
            return True
            
        elif col == 1 or col == 4 or col == 7: #bottom middle
            for i in range(1,3):
                if mat[row-i][col-1] == num:
                    return False
                if mat[row-i][col+1] == num:
                    return False
            return True

        elif col == 2 or col == 5 or col == 8: #bottom right
            for i in range(1,3):
                for j in range(1,3):
                    if mat[row-i][col-j] == num:
                        return False
            return True
```

```python
def fillMatrix(mat):
    for k in range(1, 10):
        for i in range(9):
            for j in range(9):
                if notInColumn(mat, j, k) and notInRow(mat, i, k) and notInGrid(mat, i, j, k) and auxMat[i][j] != [-1]:
                    auxMat[i][j].append(k)
```

```python
def removeFromAux(row, col, num):
    for i in range(9):
        if num in auxMat[row][i]:
            auxMat[row][i].remove(num)
    for j in range(9):
        if num in auxMat[j][col]:
            auxMat[j][col].remove(num)
    if row == 0 or row == 3 or row == 6:
        if col == 0 or col == 3 or col == 6: #top left
            for i in range(1,3):
                for j in range(1,3):
                    if num in auxMat[row+i][col+j]:
                        auxMat[row+i][col+j].remove(num)

        elif col == 1 or col == 4 or col == 7: #top middle
            for i in range(1,3):
                if num in auxMat[row+i][col+1]:
                    auxMat[row+i][col+1].remove(num)
                if num in auxMat[row+i][col-1]:
                    auxMat[row+i][col-1].remove(num)

        elif col == 2 or col == 5 or col == 8: #top right
            for i in range(1,3):
                for j in range(1,3):
                    if num in auxMat[row+i][col-j]:
                        auxMat[row+i][col-j].remove(num)

    elif row == 1 or row == 4 or row == 7:
        if col == 0 or col == 3 or col == 6: #middle left
            for j in range(1,3):
                if num in auxMat[row+1][col+j]:
                    auxMat[row+1][col+j].remove(num)
                if num in auxMat[row-1][col+j]:
                    auxMat[row-1][col+j].remove(num)
          
            
        elif col == 1 or col == 4 or col == 7: #middle middle
            if num in auxMat[row+1][col+1]:
                auxMat[row+1][col+1].remove(num)
            if num in auxMat[row-1][col+1]:
                auxMat[row-1][col+1].remove(num)
            if num in auxMat[row+1][col-1]:
                auxMat[row+1][col-1].remove(num)
            if num in auxMat[row-1][col-1]:
                auxMat[row-1][col-1].remove(num)

        elif col == 2 or col == 5 or col == 8: #middle right
            for j in range(1,3):
                if num in auxMat[row+1][col-j]:
                    auxMat[row+1][col-j].remove(num)
                if num in auxMat[row-1][col-j]:
                    auxMat[row-1][col-j].remove(num)

    elif row == 2 or row == 5 or row == 8:
        if col == 0 or col == 3 or col == 6: #bottom left
            for i in range(1,3):
                for j in range(1,3):
                    if num in auxMat[row-i][col+j]:
                        auxMat[row-i][col+j].remove(num)
            
        elif col == 1 or col == 4 or col == 7: #bottom middle
            for i in range(1,3):
                if num in auxMat[row-i][col-1]:
                    auxMat[row-i][col-1].remove(num)
                if num in auxMat[row-i][col+1]:
                    auxMat[row-i][col+1].remove(num)

        elif col == 2 or col == 5 or col == 8: #bottom right
            for i in range(1,3):
                for j in range(1,3):
                    if num in auxMat[row-i][col-j]:
                        auxMat[row-i][col-j].remove(num)
```

```python
def playSingle(mat):
    for i in range(9):
        for j in range(9):
            if len(auxMat[i][j]) == 1 and auxMat[i][j][0] != -1:
                mat[i][j] = auxMat[i][j][0]
                auxMat[i][j] = [-1]
                removeFromAux(i, j, mat[i][j])
```

```python
def notInGridA(row, col, num): 
    if row == 0 or row == 3 or row == 6:
        if col == 0 or col == 3 or col == 6: #top left
            for i in range(1,3):
                for j in range(1,3):
                    if num in auxMat[row+i][col+j]:
                        return False
            return True

        elif col == 1 or col == 4 or col == 7: #top middle
            for i in range(1,3):
                if num in auxMat[row+i][col+1]:
                    return False
                if num in auxMat[row+i][col-1]:
                    return False
            return True

        elif col == 2 or col == 5 or col == 8: #top right
            for i in range(1,3):
                for j in range(1,3):
                    if num in auxMat[row+i][col-j]:
                        return False
            return True

    elif row == 1 or row == 4 or row == 7:
        if col == 0 or col == 3 or col == 6: #middle left
            for j in range(1,3):
                if num in auxMat[row+1][col+j]:
                    return False
                if num in auxMat[row-1][col+j]:
                    return False
            return True
          
            
        elif col == 1 or col == 4 or col == 7: #middle middle
            if num in auxMat[row+1][col+1]:
                return False
            if num in auxMat[row-1][col+1]:
                return False
            if num in auxMat[row+1][col-1]:
                return False
            if num in auxMat[row-1][col-1]:
                return False
            return True

        elif col == 2 or col == 5 or col == 8: #middle right
            for j in range(1,3):
                if num in auxMat[row+1][col-j]:
                    return False
                if num in auxMat[row-1][col-j]:
                    return False
            return True

    elif row == 2 or row == 5 or row == 8:
        if col == 0 or col == 3 or col == 6: #bottom left
            for i in range(1,3):
                for j in range(1,3):
                    if num in auxMat[row-i][col+j]:
                        return False
            return True
            
        elif col == 1 or col == 4 or col == 7: #bottom middle
            for i in range(1,3):
                if num in auxMat[row-i][col-1]:
                    return False
                if num in auxMat[row-i][col+1]:
                    return False
            return True

        elif col == 2 or col == 5 or col == 8: #bottom right
            for i in range(1,3):
                for j in range(1,3):
                    if num in auxMat[row-i][col-j]:
                        return False
            return True
```

```python
def notInColumnA(row, col, num): 
    for i in range(9):
        if row != i:
            if num in auxMat[i][col]:
                return False
    return True
```

```python
def notInRowA(row, col, num):
    for i in range(9):
        if col != i:
            if num in auxMat[row][i]:
                return False
    return True
```

```python
def onlyNumberPosibility(mat, row, col):
    for k in auxMat[row][col]:
        if notInColumnA(row, col, k) and notInRowA(row, col, k) and notInGridA(row, col, k) and auxMat[row][col] != [-1]:
            mat[row][col] = k 
            removeFromAux(row, col, k)
```

```python
def singlePieceLeft(mat):
    for i in range(9):
        for j in range(9):
            if len(auxMat[i][j]) == 1 and auxMat[i][j][0] != -1:
                return True
    return False
```

```python
while not done(mat):
    while singlePieceLeft(mat):
        playSingle(mat)
        if done(mat):
            break
    for i in range(9):
        for j in range(9):
            if len(auxMat[i][j]) > 1 and auxMat[i][j] != [-1]:
                onlyNumberPosibility(mat, i, j)
    render(mat)
render(mat)
```
